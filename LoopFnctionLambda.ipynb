```python
#Write a function that uses *args to return the sum of any number of arguments.

def findsum(*args):
    total = 0
    for i in args:
        total+=i
    print("The sum is :",total)

findsum(144,20,30)
```

```python
#Write a function that uses **kwargs to print all key-value pairs passed to it.

def keyvalprint(**kwargs):
    for key,value in kwargs.items():
        print(key ,":", value)
keyvalprint(name = "DIpesh", age= 25)
```

```python
#Write a recursive function to compute the factorial of a number.
def factorial(num):
    if num==0 or num==1:
        return 1
    else:
        return num*factorial(num-1)

factorial(2)
```
